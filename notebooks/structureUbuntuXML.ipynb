```
import xml.etree.ElementTree as ET
import pandas as pd

from lxml import etree

parser = etree.XMLParser(recover=True)
root = ET.parse('Posts.xml', parser=parser).getroot()
# root = ET.parse('askubuntu_2014_posts.xml').getroot()

title = []
postId = []
PostTypeId = []
for child in root:
  attr = child.attrib
#   if 'Title' in attr:
#     title.append(attr['Title'])
#   else:
#     title.append('')
#   postId.append(attr['Id'])
  PostTypeId.append(attr['PostTypeId'])
#   print(attr)
  if(attr['PostTypeId'] == "1"):
    if 'Title' in attr:
      title.append(attr['Title'])
    else:
      title.append('')
    postId.append(attr['Id'])

df = pd.DataFrame({'title':title, 'postId':postId})
print(df.head())
print(df.shape)
```

```
rootLinks = ET.parse('PostLinks.xml').getroot()
id1 = []
id2 = []
for child in rootLinks:
  attr = child.attrib
  if((attr['LinkTypeId'] == "3") & (attr['PostId'] in df['postId'].values) & (attr['RelatedPostId'] in df['postId'].values)):
    id1.append(attr['PostId'])
    id2.append(attr['RelatedPostId'])

df_duplicate = pd.DataFrame({'id1':id1, 'id2':id2})
print(df_duplicate.head())
print(df_duplicate.shape)
```

```
df_duplicate.to_csv('duplicates.csv', sep='\t')
```
